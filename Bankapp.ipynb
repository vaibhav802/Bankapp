```python
import os
import json
import time
from getpass import getpass
#os.mkdir('Bank/')
def Main():
    print(' '*40,'* Welcome to Bank Application')
    print("1. signup")
    print("2. login")
    print('3. exit')
    choice = int(input('Enter your choice'))
    if choice==1:
        signup()
    elif choice==2:
        login()
    elif choice==3:
        exit()
    else:
        print("Invalid choice: Please select between 1-4 ")
```

```python
def signup():
    f = open('logs.txt','a')
    s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for signup"
    f.write(s)
    f.close()
    
    print(' '*50,'Welcome to signup page')
    if os.listdir('Bank/'):
        f = open('logs.txt','a')
        s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for signup"
        f.write(s)
        f.close()
        last_acc = os.listdir('Bank/')[-1].split('.')[0]
        acc = int(last_acc)+1
        d  = {'Name':input('Enter name: '),'Balance':int(input('Enter Balance: ')),'Pwd':getpass('Enter password: ')}
        f = open(f'Bank/{acc}.txt','w')
        json.dump(d,f)
        f.close()
        print('Account Created Successfully')
        print('Now you can login...')
        print('Account Number: ',acc)
        print('Name: ',d['Name'])
        print('Bal: ',d["Balance"])
        
        
        Main()
            
    else:
        f = open('logs.txt','a')
        s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for signup"
        f.write(s)
        f.close()
        d  = {'Name':input('Enter Name: '),'Balance':int(input('Enter balance')),'Pwd':getpass('Enter password: ')}
        f  = open('Bank/1001.txt','w')
        json.dump(d,f)
        f.close()
        print('Account Created Successfully')
        print('Now you can login...')
        print('Account Number: ',acc)
        print('Name: ',d['Name'])
        print('Bal: ',d["Balance"])
       
       
        Main()
    
def login():
    f = open('logs.txt','a')
    s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for login"
    f.write(s)
    f.close()
    print(' '*50,'Welcome to login page')
    acc = input('Enter bank aaccount number')
    acc = acc+'.txt'
    if acc in os.listdir('Bank/'):
        f = open(f'Bank/{acc}')
        data = json.load(f)
        
        password = getpass('Enter password: ')
        if password == data['Pwd']:
            print('Successfully login')
            print('1. Credit')
            print('2. Debit')
            print('3. Check Balance')
            print('4. Change password')
            print('5. logout')
            choice = int(input('Enter choice'))
            if choice == 1:
                def Credit():
                    f = open('logs.txt','a')
                    s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for Credit"
                    f.write(s)
                    f.close()
                    
                    print(' '*50,'Credit section')
                    amount = int(input('Enter amount to add in balance:'))
                    f = open(f'Bank/{acc}')
                    data = json.load(f)
                    data['Balance']=data['Balance']+amount
                    f = open(f'Bank/{acc}','w')
                    json.dump(data,f)
                    print('Credit successfully')
                    f.close()
                    login()
                Credit()
            elif choice==2:
                def Debit():
                    f = open('logs.txt','a')
                    s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for Debit"
                    f.write(s)
                    f.close()
                    print(' '*50,'Debit section')
                    amount= int(input('Enter amount to withdraw:'))
                    f = open(f'Bank/{acc}')
                    data = json.load(f)
                    if data['Balance']>=amount:
                        data['Balance']=data['Balance']-amount
                        f = open(f'Bank/{acc}','w')
                        data =json.dump(data,f)
                        print('Withdraw amount successfully')
                        f.close()
                        login()
                    else:
                        f = open('logs.txt','a')
                        s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for Debit"
                        f.write(s)
                        f.close()
                        print('Insufficient Balance')
                        
                
                Debit()
            elif choice == 3:
                def Check_Balance():
                    f = open('logs.txt','a')
                    s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for check Balance"
                    f.write(s)
                    f.close()
                    f = open(f'Bank/{acc}')
                    data =json.load(f)
                    bal = data['Balance']
                    print('Your Account Balance is: ',bal,'Rs')
                    f.close()
                    login()
                Check_Balance()
            
            elif choice == 4:
                def Change_password():
                    f = open('logs.txt','a')
                    s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for change password"
                    f.write(s)
                    f.close()
                    new_password = input('Enter new password:')
                    Re_enter_password = getpass('confirm password:')
                    if new_password==Re_enter_password:
                        f= open(f'Bank/{acc}')
                        data = json.load(f)
                        data['Pwd']=new_password
                        f = open('Bank/{acc}','w')
                        data = json.dump(data,f)
                        print('Password changed successfully')
                        f.close()
                        login()
                    else:
                        print("password doesn't match! Please enter carefully")
                        login()
                    
                Change_password()
            elif choice == 5:
                def logout():
                    f = open('logs.txt','a')
                    s = f"\n{time.strftime('%d-%m-%y')}\t\t{time.strftime('%H:%M:%S')}\t\t Request for logout"
                    f.write(s)
                    f.close()
                    print('successfully logout')
                    Main()
                logout()
            else:
                print('Invalid Choice! Please Choice number between 1 to 5 ')
                login()
        else:
            print('Invalid password! PLease Enter Correct password')
            Main()
    else:
        print("Account doesn't exist.")
        Main()
def exit():
    print('Thank u for using Bank Application')
    



```
